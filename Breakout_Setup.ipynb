```python
import gym
from colabgymrender.recorder import Recorder

env = gym.make('Breakout-v0')
directory = './video'
env = Recorder(env, directory)

observation = env.reset()
terminal = False
while not terminal:
  action = env.action_space.sample()
  observation, reward, terminal, info = env.step(action)

env.play()
```
